# Impute Missing Values

```python
from sklearn.neighbors import KNeighborsRegressor
#Create a function that will help you fill in missing values using KNeighbours instead of mean
```

```python
def knn_impute(df, na_target_column):
    df = df.copy()
    
    numeric_df= df.select_dtypes(np.number)
    non_na_columns = numeric_df.loc[: , numeric_df.isna().sum() == 0].columns
    
    y_train = numeric_df.loc[numeric_df[na_target_column].isna() == False, na_target_column]
    X_train =  numeric_df.loc[numeric_df[na_target_column].isna() == False, non_na_columns]
    X_test = numeric_df.loc[numeric_df[na_target_column].isna() == True, non_na_columns]
    
    knn= KNeighborsRegressor()
    knn.fit(X_train, y_train)
    
    y_pred = knn.predict(X_test)
    
    df.loc[df[na_target_column].isna() == True, na_target_column] = y_pred
    
    return df
```

```python
data2 = knn_impute(data2, 'Valencia_pressure')
```

# Feature Transformations

```python
skew_df = pd.DataFrame(data3.select_dtypes(np.number).columns, columns = ['feature'])
skew_df['skew'] = skew_df['feature'].apply(lambda feature: scipy.stats.skew(data3[feature]))
skew_df['absolute_skew'] = skew_df['skew'].apply(abs)
skew_df['skewed'] = skew_df['absolute_skew'].apply(lambda x: True if x >= 0.5 else False)
skew_df
```

```python
#correct the skewed columns by applying a log function
# apply the log (x + 1) to catter for 0 values
for column in skew_df.query('skewed == True')['feature'].values:
    data3[column] = np.log1p(data3[column])
```

```python
#check again for skewness
skew_df = pd.DataFrame(data3.select_dtypes(np.number).columns, columns = ['feature'])
skew_df['skew'] = skew_df['feature'].apply(lambda feature: scipy.stats.skew(data3[feature]))
skew_df['absolute_skew'] = skew_df['skew'].apply(abs)
skew_df['skewed'] = skew_df['absolute_skew'].apply(lambda x: True if x >= 0.5 else False)
skew_df
```
